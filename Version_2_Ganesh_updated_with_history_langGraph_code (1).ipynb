# Guardrails and Evaluator

```python
def guardrails_check(state: ChatState) -> ChatState:
    user_input = state["input"].strip()
    blocked = False
    reasons = []

    # 1. Security-sensitive keywords
    security_keywords = ["password", "otp", "credit card", "cvv", "ssn", "bank"]
    if any(word in user_input.lower() for word in security_keywords):
        blocked = True
        reasons.append("Sensitive financial/personal data detected.")

    # 2. Illegal or unethical queries
    illegal_keywords = ["steal", "hack", "bypass", "exploit", "cheat", "pirate", "crack"]
    if any(word in user_input.lower() for word in illegal_keywords):
        blocked = True
        reasons.append("Illegal or unethical intent.")

    # 3. Toxic or harmful speech
    toxic_keywords = ["kill", "hate", "bomb", "attack", "destroy"]
    if any(word in user_input.lower() for word in toxic_keywords):
        blocked = True
        reasons.append("Harmful or aggressive language.")

    # 4. Prompt injection attempts
    if any(trigger in user_input.lower() for trigger in ["ignore previous", "act as", "you are now", "system prompt"]):
        blocked = True
        reasons.append("Potential prompt injection attempt.")

    # 5. Gibberish or empty
    if len(user_input) < 2 or not re.search(r'\w+', user_input):
        blocked = True
        reasons.append("Input is empty or nonsensical.")

    # Set response if blocked
    if blocked:
        state["guardrails_response"] = "failed"
        state["response"] = f"Request blocked for safety. Reason(s): {' | '.join(reasons)}"
    
    return state
    
def evaluator(state: ChatState) -> ChatState:
    user_input = state.get("input", "").lower()
    response = state.get("response", "").lower()

    if "sorry" in response or state.get("guardrails_response") == "failed":
        state["evaluation"] = "Blocked: Guardrails triggered."
        return state

    key_phrases = [
        "plan", "feature", "benefit", "validity", "pricing", 
        "offer", "data", "unlimited", "calls", "usage"
    ]
    tone_warnings = ["sorry", "unfortunately", "cannot", "not sure", "unknown"]
    redundant_phrases = ["the plan is", "this plan", "you can", "you will get"]

    score = 0
    explanation = []

    # Relevance score
    overlap = sum(1 for word in key_phrases if word in response)
    score += overlap
    if overlap >= 3:
        explanation.append("Relevant keywords covered.")
    else:
        explanation.append("Missing some important terms.")

    # Length check
    response_length = len(response.split())
    if response_length < 10:
        explanation.append("Too short.")
        score -= 1
    elif response_length > 100:
        explanation.append("Too long, consider shortening.")
    else:
        explanation.append("Appropriate length.")
        score += 1

    # Tone check
    if any(tone in response for tone in tone_warnings):
        explanation.append("Negative or uncertain tone.")
        score -= 1
    else:
        explanation.append("Clear and confident tone.")
        score += 1

    # Repetition check
    repeats = [phrase for phrase in redundant_phrases if response.count(phrase) > 1]
    if repeats:
        explanation.append(f"Repetitive phrase(s): {', '.join(repeats)}")
        score -= 1

    # Score-based output
    if score >= 4:
        verdict = "Excellent Response"
    elif score >= 2:
        verdict = "Decent Response"
    else:
        verdict = "Needs Improvement"

    state["evaluation"] = f"{verdict} | Score: {score}/5\n" + " | ".join(explanation)
    print("The Evaluator Explaination")
    print(explanation)
    print()
    return state
    
    
```

```python

# Agent action with chat history
def agent_actions(state: ChatState) -> ChatState:
    user_input = state["input"]
    # just defined chathistory as like user input
    chat_history = state.get("chat_history", [])
    
    if len(chat_history)>4:
        chat_history = chat_history[-4:]
    print("The Chat History is")
    print()
    for msg in chat_history:
        print(f"-{msg.type}: {msg.content}")
#     print(chat_history)
    print()
    print("*"*25)
#     response = f"Processed response to: {user_input}, the plan platinum unlimited is the best one month plan with highest features and capabilities."

######### Trying a RAG Agent Dynamic URL in Agent Action #########
    # Embed summaries
    model = SentenceTransformer('all-MiniLM-L6-v2')
    # Main pipeline
    model_llm = Ollama(model="llama3.2:3b")
    # Example
    user_query = user_input + f"Chat_History : {chat_history}"

    response = rag_url_navigation(model,model_llm,user_query)
#     print("the response from primary code is")
#     print(response)
    response = str(response)
######### END #########


    chat_history.append(HumanMessage(content=user_input))
    chat_history.append(AIMessage(content=response))

    if len(chat_history)>4:
        chat_history = chat_history[-4:]
        
    state["response"] = response
    state["chat_history"] = chat_history
    return state

```

```python
# Build the LangGraph
graph = StateGraph(ChatState)

graph.add_node("guardrails", guardrails_check)
graph.add_node("agent", agent_actions)
graph.add_node("evaluator", evaluator)
graph.add_node("END", lambda state: state)

def guardrails_to_agent(state: ChatState) -> Literal["END", "agent"]:
    return "agent" if state.get('guardrails_response') != "failed" else "END"

graph.add_conditional_edges("guardrails", guardrails_to_agent)
graph.add_edge("agent", "evaluator")
graph.add_edge("evaluator", "END")

graph.set_entry_point("guardrails")
```

```python
workflow = graph.compile()
```
